### A Neuron / Perceptron 
Building block of a Neural Network

```python
## Defining sigmoid

def sigmoid(z):

    sigmoid = 1/(1 + np.exp(z))
    
    return sigmoid
```

### Layer(s) [Dense]
A set of Neurons

```python
def dense(a_in, W, B):
    
    units = W.shape[1]          # Retrieve number of cols = number of variables
    a_out = np.zeros(units)
    for j in range(units):
        w = W[:, j]
        b = B[j]
        z = np.dot(w, a_in)+b
        a_out[j] = sigmoid(z)
    
    return 
```

### Sequential 

A set of layers

```python
def sequential(x, W1, B1, W2, B2):

    a1 = dense(x, W1, B1)
    a2 = dense(a1, W2, B2)

    return a2
```
